```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Replacing the null values in Age with the mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# Replacing the null values in Embarked with the mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Dropping the Cabin column because it contains too many missing values
df.drop('Cabin', axis=1, inplace=True)
```

```python
# Checking missing values after cleaning the data
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Plotting numerical columns
plt.figure(figsize=(12,10))
plt.subplot(3,2,1)
sns.histplot(df['Age'])
plt.title("Age Distribution")

plt.subplot(3,2,2)
sns.histplot(df['Fare'])
plt.title("Fare Distribution")

plt.subplot(3,2,3)
sns.histplot(df['Pclass'])
plt.title("Passenger Class Distribution")

plt.subplot(3,2,4)
sns.histplot(df['SibSp'])
plt.title('Distribution of Siblings/Spouses Aboard')

plt.subplot(3,2,5)
sns.histplot(df['Parch'])
plt.title('Distribution of Parents/Children Aboard')

plt.tight_layout()
```

```python
# Plotting Categorical columns
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
sns.countplot(x=df['Survived'])
plt.title("Survived Count")

plt.subplot(2,2,2)
sns.countplot(x=df['Sex'])
plt.title("Gender Distribution")

plt.subplot(2,2,3)
sns.countplot(x=df['Embarked'])
plt.title("Embarkation Port Distribution")

plt.tight_layout()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='Age', y='Fare', data=df)

plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Pclass', y='Fare', data=df)

plt.title('Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(),
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')
```

```python
plt.figure(figsize=(6,6))

df['Survived'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=['Did Not Survive','Survived']
)
plt.title('Survival Distribution')
plt.ylabel('')
```
