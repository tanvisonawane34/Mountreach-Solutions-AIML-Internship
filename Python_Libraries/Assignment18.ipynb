```python
# Q10. Complete Preprocessing Pipeline

filepath = "/content/ford_car_dataset.csv"
df = pd.read_csv(filepath)

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# ==========================
# Check Missing Values
# ==========================

print("Missing Values:")
print(df.isnull().sum())

# ==========================
# Check and Remove Duplicates
# ==========================

print("\nDuplicate Rows:")
print(df.duplicated().sum())

df = df.drop_duplicates()

# ==========================
# Histogram Plots
# ==========================

plt.figure(figsize=(12,10))

# Price
plt.subplot(3,2,1)
sns.histplot(df["price"], bins=30)
plt.title("Price Distribution")

# Mileage
plt.subplot(3,2,2)
sns.histplot(df["mileage"], bins=30)
plt.title("Mileage Distribution")

# Year
plt.subplot(3,2,3)
sns.histplot(df["year"], bins=30)
plt.title("Year Distribution")

# Engine Size
plt.subplot(3,2,4)
sns.histplot(df["engineSize"], bins=30)
plt.title("Engine Size Distribution")

# MPG
plt.subplot(3,2,5)
sns.histplot(df["mpg"], bins=30)
plt.title("MPG Distribution")

plt.subplots_adjust(hspace=0.6, wspace=0.3)

# ==========================
# Count Plots
# ==========================

plt.figure(figsize=(12,8))

# Model
plt.subplot(2,2,1)
sns.countplot(data=df, x="model")
plt.title("Count of Car Models")
plt.xticks(rotation=90)

# Transmission
plt.subplot(2,2,2)
sns.countplot(data=df, x="transmission")
plt.title("Count of Transmission Types")

# Fuel Type
plt.subplot(2,2,3)
sns.countplot(data=df, x="fuelType")
plt.title("Count of Fuel Types")

plt.subplots_adjust(hspace=1.3, wspace=0.4)

# ==========================
# Correlation Heatmap
# ==========================

plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")

# ==========================
# Independent & Dependent Features
# ==========================

X = df.drop("price", axis=1)
y = df["price"]

print("Independent Features:")
print(X.columns)

print("\nDependent Feature:")
print("price")

# ==========================
# One-Hot Encoding
# ==========================

encoded_data = pd.get_dummies(df, columns=["model", "transmission", "fuelType"])

# ==========================
# Standard Scaling
# ==========================

features = ["year", "mileage", "tax", "mpg", "engineSize"]
scaler = StandardScaler()
encoded_data[features] = scaler.fit_transform(encoded_data[features])

# ==========================
# Display Processed Data
# ==========================

print("\nFirst 5 Rows of Processed Data:")
print(encoded_data.head())
```
