```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Replace missing values with 0 temporarily to calculate the mean
df['Age'] = df['Age'].fillna(0)
df['Salary'] = df['Salary'].fillna(0)
```

```python
# Plotting numerical columns
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
sns.histplot(df['Age'])
plt.title("Age")

plt.subplot(2,2,2)
sns.histplot(df['Salary'])
plt.title("Salary")

plt.subplot(2,2,3)
sns.histplot(df['Experience_Years'])
plt.title("Experience Years")

plt.subplot(2,2,4)
sns.histplot(df['Performance_Rating'])
plt.title('Performance rating')

plt.tight_layout()
```

```python
# Plotting Categorical columns
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
sns.countplot(x=df['Gender'])
plt.title("Gender")

plt.subplot(2,2,2)
sns.countplot(x=df['Department'])
plt.title("Department")

plt.subplot(2,2,3)
sns.countplot(x=df['City'])
plt.title("City")

plt.subplot(2,2,4)
sns.countplot(x=df['Work_Mode'])
plt.title("Work Mode")
plt.tight_layout()
```

```python
# Calculating the mean of non-zero values and replacing 0 values
ag = df[df['Age'] != 0]
age_mean = ag['Age'].mean()

df['Age'] = df['Age'].replace(0, age_mean)
```

```python
# Checking missing values after cleaning the data
df.isnull().sum()
```

```python
# Checking Age distribution after cleaning
sns.histplot(df['Age'])
```

```python
# Calculating the mean of non-zero values and replacing 0 values
sal = df[df['Salary'] != 0]
salary_mean = sal['Salary'].mean()

df['Salary'] = df['Salary'].replace(0, salary_mean)
```

```python
# Checking Salary distribution after cleaning
sns.histplot(df['Salary'])
```

```python
# Checking duplicate values in the dataset
df.duplicated().sum()
```
