```python
filepath = "/content/heart.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
# Seperate x and y
X = data.drop("HeartDisease", axis=1)
y = data["HeartDisease"]
```

```python
print(X)
```

```python
print(y)
```

```python
# Encoding categorical columns
X = pd.get_dummies(X, drop_first=True)
X.head()
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# predictions
y_pred = model.predict(X_test)

print("First 10 Actual Values:")
print(y_test.head(10).values)

print("First 10 Predicted Values:")
print(y_pred[:10])
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

TN, FP, FN, TP = cm.ravel()

print("TN:", TN)
print("FP:", FP)
print("FN:", FN)
print("TP:", TP)
```

```python
# Model Evaluation
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(model, "heart_model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X.columns.tolist(), "columns.pkl")

print("Model and preprocessing objects saved successfully!")
```

```python
import joblib
import pandas as pd

# Load saved objects
loaded_model = joblib.load("heart_model.pkl")
loaded_scaler = joblib.load("scaler.pkl")
columns = joblib.load("columns.pkl")

# Create sample patient data
sample_data = pd.DataFrame([{
    "Age": 55,
    "Sex": "M",
    "ChestPainType": "ATA",
    "RestingBP": 140,
    "Cholesterol": 250,
    "FastingBS": 0,
    "RestingECG": "Normal",
    "MaxHR": 150,
    "ExerciseAngina": "N",
    "Oldpeak": 1.2,
    "ST_Slope": "Up"
}])

# Encode categorical columns
sample_data = pd.get_dummies(sample_data)

# Match the columns used during training
sample_data = sample_data.reindex(columns=columns, fill_value=0)

# Scale the sample data
sample_data_scaled = loaded_scaler.transform(sample_data)

# Make prediction
prediction = loaded_model.predict(sample_data_scaled)

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Heart Disease: Yes")
else:
    print("Heart Disease: No")
```
