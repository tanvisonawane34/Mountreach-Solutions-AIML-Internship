```python
# Define Features (X) and Target (y)
X = df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
y = df["Survived"]
```

```python
# Convert Categorical Columns into Numerical Columns
X = pd.get_dummies(X, drop_first=True)
print(X.head())
print(X.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=42
)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

TN, FP, FN, TP = cm.ravel()
print("TN:", TN)
print("FP:", FP)
print("FN:", FN)
print("TP:", TP)
```

```python
# Classification Report

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# Load saved objects
loaded_model = joblib.load("titanic_model.pkl")
loaded_scaler = joblib.load("scaler.pkl")
columns = joblib.load("columns.pkl")

# Create sample passenger data
sample_data = pd.DataFrame([{
    "Pclass": 3,
    "Age": 22,
    "SibSp": 1,
    "Parch": 0,
    "Fare": 7.25,
    "Sex": "male",
    "Embarked": "S"
}])

# Encode categorical columns
sample_data = pd.get_dummies(sample_data)

# Match the columns used during training
sample_data = sample_data.reindex(columns=columns, fill_value=0)

# Scale the data
sample_data_scaled = loaded_scaler.transform(sample_data)

# Make prediction
prediction = loaded_model.predict(sample_data_scaled)
print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Survived: Yes")
else:
    print("Survived: No")
```
