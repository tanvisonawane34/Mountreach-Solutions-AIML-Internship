```python
# Q5
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.subplot(2,2,1)
sns.histplot(df['age'])
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")


plt.subplot(2,2,2)
sns.histplot(df['bmi'])
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Count")


plt.subplot(2,2,3)
sns.histplot(df['children'])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Count")


plt.subplot(2,2,4)
sns.histplot(df['expenses'])
plt.title("Charges Distribution")
plt.xlabel("Charges")
plt.ylabel("Count")
plt.tight_layout()

```

```python
# Q6
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
sns.countplot(x=df['sex'])
plt.title("Sex Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.countplot(x=df['smoker'])
plt.title("Smoker Distribution")
plt.xlabel("Smoker")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.countplot(x=df['region'])
plt.title("Region Distribution")
plt.xlabel("Region")
plt.ylabel("Count")

plt.tight_layout()
```

```python
# Q7
correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
```

```python
# Q8
df['expenses'].mean()
df['expenses'].max()
df['expenses'].min()

df.groupby('smoker')['expenses'].mean()
```

```python
# Q9
sns.boxplot(x='smoker', y='expenses', hue='sex', data=df)

plt.title("Expenses by Smoker and Sex")
plt.xlabel("Smoker")
plt.ylabel("Expenses")
```
