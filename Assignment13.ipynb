```python
import numpy as np
arr = np.random.rand(10)
print("10 random nos between 0 & 1 :")
print(arr)

arr2 = np.random.randn(3,3)
print("\nA 3x3 matrix of random nos :")
print(arr2)

arr3 = np.random.randint(10,51,(4,5))
print("\n2D Array of random integers between 10 to 50 :")
print(arr3)
```

```python
import numpy as np
v1 = np.array([2,4,6,8])
v2 = np.array([1,3,5,7])

print("Vector 1 :")
print(v1)

print("\nVector 2 :")
print(v2)

print("\nAddition :")
print(v1 + v2)

print("\nSubtraction :")
print(v1 - v2)

print("\nElement-wise Multiplication :")
print(v1 * v2)

print("\nDot Product:")
print(np.dot(v1, v2))
```

```python
import numpy as np
arr = np.random.randint(1,101,(4,4))
print("4X4 matrix of random integers :")
print(arr)

print("\nShape :",arr.shape)
print("Dimension :",arr.ndim)
print("Size :",arr.size)
print("Datatype :",arr.dtype)
print("Minimum value :",arr.min())
print("Maximum value :",arr.max())
```

```python
import numpy as np
arr = np.random.randint(1,51,(20))
print("1D array :")
print(arr)

matrix = arr.reshape(4,5)
print("\n4x5 Matrix :")
print(matrix)

print("\nSum:", matrix.sum())
print("Mean:", matrix.mean())
print("Standard Deviation:", matrix.std())

print("\nMaximum Value in Each Row:")
print(matrix.max(axis=1))
```

```python
# Import numpy
import numpy as np

# Exception handling
try :

  # Ask the user to enter numbers to generate
  num = int(input("Enter numbers to Generate : "))
  # To print random nos between 1-100
  arr = np.random.randint(10,101,(num))
  print("Generated array : ")
  print(arr)

  # To print the math functions
  print("\nMean :",arr.mean())
  print("Median :",np.median(arr))
  print("Standard Deviation :",arr.std())
  print("Minimum value :",arr.min())
  print("Maximum value :",arr.max())

  # Converting the array into 2D array
  if num % 2 == 0:
      matrix = arr.reshape(2, num // 2)

      print("\n2D Array:")
      print(matrix)

      print("\nRow-wise Sum:")
      print(matrix.sum(axis=1))

  else:
      print("\nCannot reshape into a 2D array because the number of elements is odd.")

# Except block
except ValueError :
  print("Please enter a valid number :")

```
